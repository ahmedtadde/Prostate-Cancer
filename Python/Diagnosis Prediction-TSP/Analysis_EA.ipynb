```python
###### SAME COMMENTTS ACCROSS ALL NOTEBOOKS [ CHECK Analysis_AA from DIAGNOSIS PREDICTION-TSP]

import numpy as np
from scipy import stats
import pandas as pd
from sklearn import cross_validation
from sklearn.cross_validation import train_test_split
from sklearn.cross_validation import StratifiedKFold
from sklearn import metrics
from sklearn.preprocessing import label_binarize
from joblib import Parallel, delayed  
import multiprocessing
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier



repeated = range(200) 
num_cores = multiprocessing.cpu_count()


def get_folds_scores(model, metric, seed):
    

    
    if model == "svm":
        model = SVC(kernel='rbf', probability = True)
    elif model =="knn":
        model = KNeighborsClassifier()
    elif model == "bayes":
        model = GaussianNB()
    elif model == "rf":
        model = RandomForestClassifier()
    else :
        return ("NO MODEL NAME GIVEN (CHOOSE EITHER svm, nb, knn, or rf)")
    

    
    cv = StratifiedKFold(y, n_folds= 5, shuffle= True, random_state = seed )
    scores = cross_validation.cross_val_score(model, X,y, cv=cv, scoring=metric)

    return scores


X = pd.read_csv('X_ea.csv')
y = pd.read_csv('Y_ea.csv')
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=666)
y = label_binarize(y, classes=['Normal', 'Cancer'], neg_label = -1).flatten()
y_train = label_binarize(y_train, classes=['Normal', 'Cancer'], neg_label = -1).flatten()
y_test = label_binarize(y_test, classes=['Normal', 'Cancer'], neg_label = -1).flatten()

print X.shape
print y.shape
```

```python
print
print "For SVM: "
print

svm = SVC(kernel='rbf', probability = True)
svm.fit(X,y)
y_pred = svm.predict_proba(X_test)
y_pred2 = svm.predict(X_test)
probs = [array[1] for array in y_pred]
roc_analysis = metrics.roc_curve(y_test, probs, pos_label=1)
auc = metrics.roc_auc_score(y_test, probs, average='macro')
confusion_matrix = metrics.confusion_matrix(y_test, y_pred2)
report = metrics.classification_report(y_test, y_pred2, target_names=["Normal", "Cancer"])

print 
print
print "Confusion Matrix: "
print 
print confusion_matrix
print 
print "Classification Report: "
print 
print report
print
print
print
print
print "5-fold Cross Validation repeated 200 (1000 folds'scores): "
print

Scores = Parallel(n_jobs=num_cores)(delayed(get_folds_scores)("svm","roc_auc", seed) for seed in repeated)
Scores = np.concatenate( Scores, axis=0 )

CI = stats.norm.interval(0.95, loc=Scores.mean(), scale=Scores.std()/np.sqrt(len(Scores)))
print("AUC : %0.4f +/- %0.4f") % (Scores.mean(), CI[1]-Scores.mean())
print

Scores = Parallel(n_jobs=num_cores)(delayed(get_folds_scores)("svm","accuracy", seed) for seed in repeated)
Scores = np.concatenate( Scores, axis=0 )

CI = stats.norm.interval(0.95, loc=Scores.mean(), scale=Scores.std()/np.sqrt(len(Scores)))
print("Accuracy : %0.4f +/- %0.4f") % (Scores.mean(), CI[1]-Scores.mean())
print
```
